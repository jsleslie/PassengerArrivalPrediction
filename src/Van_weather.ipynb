```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
```

```python
base_url = "https://weather.gc.ca/forecast/hourly/bc-74_metric_e.html"
response = requests.get(base_url)
soup = BeautifulSoup(response.text, 'html.parser') # Parse with Beautiful Soup
```

```python
table = soup.find("table", {"class": "table table-striped table-hover table-condensed wxo-media"})
table_rows = table.find_all('tr')
```

```python
l = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    l.append(row)
```

```python
l = [x for x in l if x != []]
forecast = pd.DataFrame(l, columns=["Time", "Temp", "Cond", "Precip", "Wind"])
```

```python
forecast['Time'] = pd.to_datetime(forecast['Time'])
forecast['Time'] = forecast['Time'].dt.time
forecast['Temp'] = pd.to_numeric(forecast['Temp'])
```

```python
forecast['Cond'].replace(regex=True,inplace=True,to_replace=r'\n',value=r'')
forecast['Wind'].replace(regex=True,inplace=True,to_replace=r'\n',value=r'')
forecast['Wind'].replace(regex=True,inplace=True,to_replace=r'gust.*',value=r'')
forecast['Wind'].replace(regex=True,inplace=True,to_replace=r'(?![0-9]).',value=r'')
```
